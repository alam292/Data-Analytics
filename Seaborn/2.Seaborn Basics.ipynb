# Seaborn Basics

#### Seaborn provides simple and attractive visualization tools built on top of Matplotlib

### Figure Size

Figure size controls the width and height of the graph.

**Syntax : plt.figure(figsize=(width, height))**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.figure(figsize = (8,5))
sns.scatterplot(x = 'total_bill', y = 'tip' , data = tips)
plt.show()
```

### Titles and Labels

Used to give information about the graph.

```python
plt.title("Graph Title")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
```

```python
plt.figure(figsize = (5,3))

sns.scatterplot(x = 'total_bill', y  = 'tip', data = tips)

plt.title('Total Bills vs Tip')
plt.xlabel('Total Bills')
plt.ylabel('Tip Amount')

plt.show()
```

### Themes and Styles

Seaborn provides built-in themes for attractive plots.

#### 1. whitegrid

White background with grid lines.

**sns.set_style("whitegrid")**

Useful for:

Statistical plots

Easy value comparison

#### 2. darkgrid

Dark background with grid lines.

**sns.set_style("darkgrid")**

Useful for:

Better contrast

Professional appearance

#### 3. white

Plain white background without grids.

**sns.set_style("white")**

Useful for:

Clean visualization

Minimal design

#### 4. dark

Dark background without grids.

**sns.set_style("dark")**

Useful for:

Dark-themed plots

Presentation visuals

#### 5. ticks

Shows axis ticks with simple styling.

**sns.set_style("ticks")**

Useful for:

Scientific graphs

Publication-style charts

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_style("darkgrid")

sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.title("Darkgrid Style")

plt.show()
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.title("Whitegrid Style")

plt.show()
```

### Color Palettes

Color palettes improve the appearance of plots.

**Syntax : sns.color_palette()**

This displays Seaborn’s default colors.

**Viewing Different Palettes**

Default Palette --- > sns.color_palette()

Deep Palette --- > sns.color_palette("deep")

Pastel Palette --- > sns.color_palette("pastel")

Bright Palette --- > sns.color_palette("bright")

**Using Palette in Plot** sns.barplot(x="day", y="total_bill", data=tips, palette="pastel")

**Common Palettes**

| Palette | Use                         |
| ------- | --------------------------- |
| deep    | Default professional colors |
| pastel  | Soft colors                 |
| bright  | Vibrant colors              |
| dark    | Dark shades                 |
| muted   | Less intense colors         |


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

plt.figure(figsize=(8,5))

sns.set_style("whitegrid")

sns.barplot(
    x="day",
    y="total_bill",
    data=tips,
    palette="deep"
)

plt.title("Average Total Bill Per Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")

plt.show()
```
