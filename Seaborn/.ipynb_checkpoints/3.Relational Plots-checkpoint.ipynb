# Relational Plot 

Relational plots are used to understand the relationship between variables in a dataset.

In Seaborn, the main relational plots are:

Scatter Plot

Line Plot


# Scatter Plot

A scatter plot shows the relationship between two numerical variables.

**Syntax : sns.scatterplot(x='x', y='y', data=df)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
sns.scatterplot(x = 'total_bill' , y = "tip" , data = tips)
plt.show()
```

```python
# Explanation
# x → Independent variable
# y → Dependent variable
# Each point represents one observation
```

### Topics in Scatter Plot

#### 1. hue

Used to add colors based on categories.

```python
# Explanation
# Male and Female points get different colors.
# Helps compare categories visually.


sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue= 'sex')
```

#### 2. style

Changes marker styles based on categories.

```python
# Explanation
# Different marker shapes for smokers and non-smokers.

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips , style = 'smoker')
```

### 3. size

Changes marker size according to a variable.

```python
# Explanation
# Larger family/group size → Larger dots.

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, size = 'size')
```

### 4. markers

Custom marker shapes.

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips,style = 'sex', marker = True)
```

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="day",
    style="smoker",
    size="size",
    data=tips
)

plt.title("Scatter Plot Example")
plt.show()
```
