### Boolean Indexing

used for filtering data using condition

```python
a
```

##### Filter Value

```python
print(a>25)
```

```python
print(a > 10)
```

```python
print(a >= 10)
```

```python
print(a<10)
```

```python
even = a[a%2 == 0]
print(even)
```

```python
even = a%2 == 0
print(even)
```

### Fancy Indexing 

Access using list of index fancy indexing means using list of index.

```python
x = n.array([31,41,51,61,71,81])
print(x)
```

```python
x[2]
```

```python
print(x[[4,3,0,1]])
```

```python
print(x[4,3,0,1])
```

```python
print(x[[3,0,2]])
```

```python
y = n.array([[3,6,9,12,15],
            [18,21,24,27,30],
            [33,36,39,42,45]])
print(y)
```

```python
print(y[[0,1]])
```

```python
print(y[[0,2]])
```

```python
print(y[:,[0,2]])
```

```python
print(y[:,[3,4]])
```

```python
y
```

```python
print(y[[0, 1, 2], [2, 1, 0]])
```

# Slicing

Slicing means gettting a range of elements

**Syntax :-  arr[start : end : step]**

```python
a
```

```python
print(a[0:3])
```

```python
print(a[0:5:1])
```

```python
print(a[2:5:1])
```

```python
a[0:5:2]
```

```python
print(a[4:0:-1])
```

```python
a[::-1]
```
