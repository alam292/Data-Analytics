# Box Plot

### What is a Box Plot?

A Box Plot (or Box-and-Whisker Plot) is used to visualize:

Data distribution

Spread of data

Median

Quartiles

Outliers

**It is very useful in:**

Data Analytics

Statistics

Exploratory Data Analysis (EDA)

### Main Purpose of Box Plot

Box plots help in:-

Detecting outliers

Understanding spread

Comparing distributions

Finding skewness

### Understanding Box Plot Structure

#### Minimum ---- Q1 ---- Median ---- Q3 ---- Maximum

### Creating a Box Plot

**Syntax : plt.boxplot(data)**

```python
import matplotlib.pyplot as plt
```

```python
data = [10,12,15,18,22,30,39,100]

plt.boxplot(data)

plt.title('Box PLot')
plt.show()
```

### Understanding Quartiles and Median

A box plot mainly contains:

| Component   | Meaning                |
| ----------- | ---------------------- |
| Minimum     | Smallest value         |
| Q1          | First Quartile (25%)   |
| Median (Q2) | Middle value (50%)     |
| Q3          | Third Quartile (75%)   |
| Maximum     | Largest value          |
| Outliers    | Unusual extreme values |


### Quartiles

Quartiles divide data into 4 equal parts.

| Quartile | Meaning        |
| -------- | -------------- |
| Q1       | 25% data below |
| Q2       | Median         |
| Q3       | 75% data below |


### Median in Box Plot

The line inside the box represents the median.

Median = middle value of the dataset.

### Outlier Detection

Outliers are values very far from other values.

In box plots:

     Outliers appear as separate dots.

### Example with Outlier

```python
marks = [45,50,52,55,60,62,65,70,150]       # 150 appears as an outlier.

plt.boxplot(marks)

plt.title('Students Marks')
plt.show()
```

### Multiple Box Plots

We can compare multiple datasets.

```python
class_A = [45, 50, 55, 60, 65]
class_B = [40, 42, 48, 52, 58]

plt.boxplot([class_A, class_B])

plt.title('Class Performance Comparison')
plt.show()
```

## Customizing Box Plot

**Important Parameters Summary:-**

| Parameter           | Purpose                     |
| ------------------- | --------------------------- |
| `patch_artist=True` | Fills box color             |
| `showmeans=True`    | Displays mean               |
| `vert=False`        | Creates horizontal box plot |


#### Change Box Color

```python
plt.boxplot(data, patch_artist=True)
plt.show()
```

#### Show Mean

```python
plt.boxplot(data, showmeans=True)
plt.show()
```

#### Horizontal Box Plot

```python
plt.boxplot(data, vert=False)
plt.show()
```
