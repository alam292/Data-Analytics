# Customization


### Customization helps make graphs:

More attractive

Easier to understand

More professional

**Matplotlib provides many customization options such as:**

Colors

Line styles

Markers

Transparency

Line width

## Important Parameters Summary

| Parameter   | Purpose                 |
| ----------- | ----------------------- |
| `color`     | Changes line color      |
| `linestyle` | Changes line style      |
| `marker`    | Adds symbols on points  |
| `alpha`     | Controls transparency   |
| `linewidth` | Controls line thickness |


```python
import matplotlib.pyplot as plt
```

### Basic Customized Plot

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat' , 'Sun']
y = [8, 9, 11, 9, 7, 15, 16]

plt.plot(x,y)
```

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y = [8, 9, 11, 9, 7, 15, 16]

plt.plot(x,y , color = 'red', linestyle = '-.' , marker = '*' )
plt.title("Customized Line Plot")
```

### Colors

The color parameter changes line color.

**Syntax : plt.plot(x, y, color='red')**

| Color Name | Short Form |
| ---------- | ---------- |
| red        | r          |
| blue       | b          |
| green      | g          |
| black      | k          |
| yellow     | y          |
| cyan       | c          |
| magenta    | m          |


```python
plt.plot(x,y, color = 'magenta')
```

### Line Styles

The linestyle parameter changes line appearance.

**Syntax : plt.plot(x, y, linestyle='--')**

| Style  | Meaning       |
| ------ | ------------- |
| `'-'`  | Solid line    |
| `'--'` | Dashed line   |
| `':'`  | Dotted line   |
| `'-.'` | Dash-dot line |


```python
plt.plot(x,y, linestyle = '-.')
```

### Markers

Markers highlight data points.

**Syntax : plt.plot(x, y, marker='o')**

| Marker | Shape    |
| ------ | -------- |
| `'o'`  | Circle   |
| `'*'`  | Star     |
| `'s'`  | Square   |
| `'^'`  | Triangle |
| `'+'`  | Plus     |
| `'x'`  | Cross    |


```python
plt.plot(x,y, marker = 's')
```

### Marker Color 

The marker color changes the color of the marker (data point symbols).

We mainly use:

| Parameter                  | Purpose                |
| -------------------------- | ---------------------- |
| `markerfacecolor` or `mfc` | Inside color of marker |
| `markeredgecolor` or `mec` | Border color of marker |


```python
plt.plot(x,y , marker = 's' , mfc = 'red')
```

```python
plt.plot(x,y , marker = 's' , mec = 'red')
```

### Marker Size

We can also increase marker size using:

markersize = 10

```python
plt.plot(x,y, marker = 'o' , markersize = 15)
```

### Transparency (alpha)

The alpha parameter controls transparency.

**Range**

0 → Fully transparent

1 → Fully visible

**Syntax : plt.plot(x, y, alpha=0.5)**

```python
plt.plot(x,y, color = 'r' , alpha = 0.3)
```

### Line Width

The linewidth parameter changes line thickness.

**Syntax : plt.plot(x, y, linewidth=3)**

```python
plt.plot (x,y , linewidth = 5)
```

### Complete Customized Example

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,30,25]

plt.figure(figsize=(8,5))

plt.plot(
    x,
    y,
    color='red',
    linestyle='--',
    marker='o',
    linewidth=3,
    alpha=0.4,
    label='Sales'
)

plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.legend()
plt.grid()

plt.show()
```
