# Scatter Plot

### What is a Scatter Plot?

A Scatter Plot is used to show the relationship between two variables.

Each point in the graph represents one observation.

**It helps in:**

Finding correlation

Detecting trends

Identifying outliers

Understanding data distribution

**Important Parameters Summary :-**

| Parameter | Purpose      |
| --------- | ------------ |
| `s`       | Marker size  |
| `c`       | Marker color |
| `alpha`   | Transparency |
| `cmap`    | Color theme  |


### Relationship Between Variables

Scatter plots are mainly used to check:

| Relationship Type    | Meaning                     |
| -------------------- | --------------------------- |
| Positive Correlation | As X increases, Y increases |
| Negative Correlation | As X increases, Y decreases |
| No Correlation       | No clear relationship       |


### Creating a Scatter Plot

**Syntax : plt.scatter(x, y)**

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,14,56,83,78,99]

plt.scatter(x,y)

plt.title('Study_Hours V/S Marks')
plt.xlabel('Study hours')
plt.ylabel('Marks')

plt.show()
```

### Marker Size

The ***s*** parameter controls marker size.

**Syntax  : plt.scatter(x, y, s=100)**

```python
plt.scatter(x,y , s = 250)
```

### Color Mapping

The ***c*** parameter changes marker colors.

**Single Color : plt.scatter(x, y, c='red')**

```python
plt.scatter(x,y , c  = 'red')
```

##### Multiple Colors

```python
colors = ['red', 'blue', 'green', 'orange', 'purple', 'black','magenta']

plt.scatter(x,y , c  = colors)
```

### Using Colormap

plt.scatter(x, y, c=y, cmap='viridis')

Here:

c=y → colors based on y-values

cmap → color theme

```python
plt.scatter(x, y, c=y, cmap='viridis')
```

### Complete Scatter Plot Example

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [10,20,15,30,25,40,35]

plt.figure(figsize=(8,5))

plt.scatter(
    x,
    y,
    s=150,
    c='red',
    alpha=0.6
)

plt.title("Sales Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid()

plt.show()
```

### Difference Between Line Plot and Scatter Plot

| Line Plot                 | Scatter Plot                         |
| ------------------------- | ------------------------------------ |
| Points connected by lines | Individual points only               |
| Shows trends over time    | Shows relationship between variables |
| Continuous data           | Correlation analysis                 |
