# Bar Charts

Bar charts are used to compare values between different categories.

They are widely used in:

Sales analysis

Survey results

Product comparison

Business reports

**Types of Bar Charts :-**

1. Vertical Bar Chart :- Uses plt.bar()

2. Horizontal Bar Chart :- Uses plt.barh()

```python
import matplotlib.pyplot as plt
```

### 1.Vertical Bar Chart

**Syntax : plt.bar(categories, values)**

```python
categories = ['A','B','C','D']
values = [10,25,15,38]

plt.bar(categories , values ,label = categories , color = 'g')
plt.title('Sales by Product')
plt.xlabel('Categories ---------->')
plt.ylabel('Values ------------>')

plt.legend()
plt.show()
```

### 2. Horizontal Bar Chart

**Syntax : plt.barh(categories, values)**

```python
categories = ['Jan','Feb','Mar','Apr','May']
sales = [700, 550, 630, 799, 900]

plt.barh(categories, sales , color = 'r' , label = categories)

plt.legend()
```

### Bar Width

The width parameter changes bar thickness.

**Syntax : plt.bar(categories, values, width=0.5)**

```python
plt.bar(categories , sales , width = 0.6 )
```

### Color Customization

We can customize bar colors using color.

##### single color 

```python
plt.bar (categories , sales, color = 'y')
```

##### multiple color 

```python
import matplotlib.pyplot as plt

products = ['Laptop', 'Mobile', 'Tablet']
sales = [50, 80, 40]

plt.bar(
    products,
    sales,
    color=['blue', 'green', 'red']
)

plt.title("Product Sales")

plt.show()
```

### Grouped Bar Chart

Grouped bar charts compare multiple datasets side by side.

```python
import matplotlib.pyplot as plt
import numpy as np

products = ['A', 'B', 'C']

sales_2024 = [50, 60, 70]
sales_2025 = [55, 65, 80]

x = np.arange(len(products))

width = 0.35

plt.bar(x - width/2, sales_2024, width, label='2024')
plt.bar(x + width/2, sales_2025, width, label='2025')

plt.xticks(x, products)

plt.title("Yearly Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales")

plt.legend()

plt.show()
```
