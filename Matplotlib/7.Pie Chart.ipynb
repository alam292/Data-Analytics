# Pie Chart

### What is a Pie Chart?

A Pie Chart is used to represent data in the form of percentages.

**It shows:**

  Parts of a whole

  Contribution of categories

  Percentage distribution

Each slice represents a category.

### Percentage Representation

Pie charts are mainly used when:

 Total value is important

 We want percentage comparison

 Data has categories

## Creating a Pie Chart

**Syntax : plt.pie(values, labels=labels, autopct='%1.1f%%')**

#### Parameters Explanation


| Parameter | Purpose              |
| --------- | -------------------- |
| `values`  | Numerical data       |
| `labels`  | Adds category names  |
| `autopct` | Shows percentages    |
| `explode` | Separates slices     |
| `shadow`  | Adds shadow          |
| `colors`  | Changes slice colors |



```python
import matplotlib.pyplot as plt
```

```python
values = [40,30,25,5]
labels = ['python', 'java' , 'c++' ,'kotlin']

plt.pie(values, labels = labels , autopct = '%1.1f%%')

plt.title('Programming Langauge Usage')
plt.show()
```

### autopct='%1.1f%%'

Used to display percentages.

**Meaning**
| Symbol  | Meaning             |
| ------- | ------------------- |
| `%1.1f` | 1 decimal float     |
| `%%`    | Displays `%` symbol |


### Explode in Pie Chart

The explode parameter separates slices from the center.

Used to highlight a category.

**Syntax : explode = [0, 0.2, 0, 0]**

Here: 0.2 → second slice moves outward

```python
import matplotlib.pyplot as plt

values = [10,30,25,45]
labels = ['Java','Python','C++','SQL']
explode = [0.11,0.15,0.1,0]

plt.pie(values, labels = labels ,  explode = explode, autopct = '%1.1f%%')
plt.show()
```

### Shadow Effect

The shadow=True parameter adds shadow.

**Syntax : shadow=True**

```python
plt.pie(values ,  labels = labels , autopct = '%1.1f%%', shadow = True)
plt.show()
```

### Colors in Pie Chart

The colors parameter customizes slice colors.

```python
colors = ['red','blue', 'yellow','green']

plt.pie(values , labels = labels , colors = colors , autopct = '%1.1f%%')
plt.show()
```

## Complete Pie Chart Example

```python
import matplotlib.pyplot as plt

values = [40, 30, 20, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']

explode = [0, 0.1, 0.07, 0]

colors = ['gold', 'skyblue', 'lightgreen', 'orange']

plt.figure(figsize=(7,7))

plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    colors=colors
)

plt.title("Programming Language Usage")

plt.show()
```
