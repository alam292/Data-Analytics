# Basic Plotting in Matplotlib

### Line Plot

A Line Plot is used to display data points connected by straight lines.

**It is mainly used to:**

Show trends over time

Compare values

Analyze growth or decline

### Creating a simple Line Chart

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title('Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales')
```

### 
| Function   | Purpose                |
| ---------- | ---------------------- |
| `plot()`   | Creates the line graph |
| `title()`  | Adds title to graph    |
| `xlabel()` | Adds label to X-axis   |
| `ylabel()` | Adds label to Y-axis   |
| `show()`   | Displays graph         |


### Multiple Lines in One Graph

We can compare multiple datasets in a single graph.

```python
months = ['Jan','Feb','Mar','Apr','May']

sales_A = [10,20,30,40,50]
sales_B = [15,25,35,30,45]

plt.plot(months,sales_A  , label = 'Product A')
plt.plot(months, sales_B , label = 'Product B')

plt.title('Monthly- wise sales comparision')
plt.xlabel('Months')
plt.ylabel('Sales')

plt.legend()

plt.show()
```

### legend() function

The legend() function displays labels for multiple lines.

Without legend(), we cannot identify which line belongs to which dataset.

**Syntax: plt.legend( )**

### grid() function

The grid() function adds grid lines to the graph.

It improves readability.



```python
x = [1,2,3,4,5]
y = [10,20,40,44,30]

plt.plot(x,y)

plt.title('Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')

plt.grid()
plt.show()
```

### Complete Example Using All Functions

```python
import matplotlib.pyplot as plt

month = ['Jan','Feb','Mar','Apr','May']

sales_A = [10,22,30,25,50]
sales_B = [15,25,35,30,45]

plt.plot(month, sales_A, label = "Product A")
plt.plot(month, sales_B, label = "Product B")

plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.legend()
plt.grid()

plt.show()
```
