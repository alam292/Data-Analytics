# Subplot

### What are Subplots?

Subplots allow multiple graphs inside a single figure.

**They are useful for:**

Comparing multiple charts

Dashboard-style visualization

Data analysis reports

**Advantages of Subplots:-**

Compare multiple charts together

Save space

Professional dashboard appearance

Better data analysis

### Creating Subplots

**Syntax : fig, ax = plt.subplots(2,2)**

**means:** 2 rows + 2 columns = Total plots = 4

### Understanding fig and ax

| Variable | Meaning                 |
| -------- | ----------------------- |
| `fig`    | Complete figure/window  |
| `ax`     | Axes (individual plots) |


## Important Functions Summary

| Function         | Purpose                |
| ---------------- | ---------------------- |
| `subplots()`     | Creates multiple plots |
| `ax[row,col]`    | Accesses subplot       |
| `figsize`        | Controls figure size   |
| `tight_layout()` | Adjusts spacing        |


```python
import matplotlib.pyplot as plt
```

```python
fig , ax = plt.subplots(2,2)
plt.show()
```

### Plotting in Subplots

Each subplot is accessed using row and column indexing.

```python
x = [1,2,3,4]
y = [10,20,15,25]

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x,y)
ax[0,0].set_title('Line Plot')

ax[0,1].bar(x,y)
ax[0,1].set_title('Bar Chart')

ax[1,0].scatter(x,y)
ax[1,0].set_title('Scatter Plot')

ax[1,1].boxplot(x,y)
ax[1,1].set_title('Box Plot')

plt.show()
```

### Accessing Subplots

| Position                  | Access    |
| ------------------------- | --------- |
| First row, first column   | `ax[0,0]` |
| First row, second column  | `ax[0,1]` |
| Second row, first column  | `ax[1,0]` |
| Second row, second column | `ax[1,1]` |


### Single Row Subplots

### Figure Size with Subplots

```python
fig, ax = plt.subplots(2,2, figsize=(10,8))
```

### tight_layout()

Used to automatically adjust spacing between subplots.

Without it:

Titles and labels may overlap.

**Syntax : plt.tight_layout()**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

fig, ax = plt.subplots(2,2, figsize=(8,6))

ax[0,0].plot(x, y)
ax[0,1].bar(x, y)
ax[1,0].scatter(x, y)
ax[1,1].hist(y)

plt.tight_layout()

plt.show()
```

### Complete Example

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,30,25]

fig, ax = plt.subplots(2,2, figsize=(10,8))

ax[0,0].plot(x, y)
ax[0,0].set_title("Line Plot")

ax[0,1].bar(x, y)
ax[0,1].set_title("Bar Chart")

ax[1,0].scatter(x, y)
ax[1,0].set_title("Scatter Plot")

ax[1,1].pie(
    [30,40,20,10],
    labels=['A','B','C','D'],
    autopct='%1.1f%%'
)
ax[1,1].set_title("Pie Chart")

plt.tight_layout()

plt.show()
```
