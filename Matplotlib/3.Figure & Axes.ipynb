# Figure and Axes

### Understanding Figure and Axes

In Matplotlib:

Figure → The complete window or canvas

Axes → The actual graph/chart area inside the figure

| Component | Meaning                     |
| --------- | --------------------------- |
| Figure    | Entire notebook page        |
| Axes      | The graph drawn on the page |


# 1.Figure in Matplotlib

A Figure is the top-level container for all plot elements.

**It can contain:**

One graph

Multiple graphs

Titles
Labels

```python
import matplotlib.pyplot as plt
```

#### plt.figure() 

Used to create a new figure.

```python
plt.figure()

x = [1,2,3,4,5]
y = [14,20,45,39,60]

plt.plot(x,y)
```

### Figure Size

We can control graph size using figsize.

**Syntax : plt.figure(figsize = (8,5))**

Here:

8 → Width (inches)

5 → Height (inches)

```python
plt.figure(figsize = (8,5))
plt.plot(x,y)

plt.title('Sales Trends')
```

```python
plt.figure(figsize = (8,5) , facecolor = 'skyblue')
plt.plot(x,y)

plt.title('Sales Trends')
```

# 2.Axes in Matplotlib

Axes represent:

X-axis

Y-axis

Plotting area

This is where the actual graph is drawn.

### fig, ax = plt.subplots()

This is the most recommended and professional way to create plots.

**Syntax : fig, ax = plt.subplots( )**

Here:

fig → Figure object

ax → Axes object


```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (8,5))

x = [1,2,3,4,5]
y = [10,20,15,30,25]

ax.plot(x, y)

ax.set_title("Sales Trend")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")

plt.show()
```
