### Dirty Data :-

data that is missing wrong , inconsistent or invalid. 

### What is Dirty Data ?

**Example :-**

**Missing values :-** None, " ", NaN

**Wrong types :-** "abc" , instead of 104

**Outlier :-** -9999, 10000

**Invalid dates :-** '32-13-2024'

**Duplicate records**

### **Rules :-** Never trust data . Always clean it.

#### Handling Wrong Data Types

```python
data = ['10','20','30','abc',None,'30.2','50']
clean = []

for x in data:
    try:
        clean.append(int(x))
    except (ValueError, TypeError):
        pass
print(clean)
```

#### Handling Missing Values

```python
data = [10,None,20,' ','20',29]
clean = [x for x in data if x not in (None,' ','20')]
print(clean)
```

#### Handling Division by Zero

```python
profit = 100
revenue = 0
try:
    margin = profit/revenue
except ZeroDivisionError:
    margin = 0
print(margin)
```

#### Handling Dirty Dictionary Data

```python
record = {'name' : 'Aman'}
age = record.get('age',10)
print(age)
```

#### Handling Dirty Dates

```python
from datetime import datetime

dates = ['21-09-2024','32-01-2024']

for d in dates:
    try:
        print(datetime.strptime(d,"%d-%m-%y"))
    except ValueError:
        print('Invalid date skipped')
```

#### Handling Dirty CSV Data

```python
import csv

with open('sales.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        try:
            print(int(row[1]))
        except:
            print('Bad row skipped')
```
