### 1. ValueError (wrong value type) :-

occurs when a value is correct type but invalid value.

```python
try:
    x = int('abc')
except ValueError:
    print('ValueError : Invalid number')
```

### 2. ZeroDivisionError (Divide by zero) :-

```python
try:
    print(10/0)
except ZeroDivisionError:
    print('ZeroDivisionError : cannot divide by zero' )
```

### 3. FileNotFoundError (File missing) :-

```python
try:
    f = open('data.txt')
except FileNotFoundError:
    print('FileNotFoundError : File does not exit')
```

### 4. TypeError(wrong operation) :-

```python
try:
    print(10 + '5')
except TypeError:
    print('TypeError : cannot add int and string')
```

### 5. KeyError (missing dictionary key) :-

```python
data = {'name' : 'Aman'}
try:
    print(data['age'])
except KeyError:
    print('KeyError : key not found')
```
