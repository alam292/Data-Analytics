### What is an Exception?

An exception is an error that occurs during program execution.

```python
print(10/0)
```

### Why Exception Handling ?

Prevent program crash 

Handle wrong input

Clean dirty data

Essential for Data Analytics

### Exception Handling allows programs to handle runtime errors gracefully without crashing.

**try :-** (test) do the work --> Always first

**except :-** handle mistake --> If error occurs

**else :-** success message --> If no error

**finally :-** cleanup (always) --> Aalways

#### Basic Structure :- 

### Example :-

```python
try:
    a = int(input('Enter a number :'))
    b = int(input('Enter a number :'))
    result = a/b
except ValueError:
    print('Please enter number only')
except ZeroDivisionError:
    print('Can''t divide by zero')
else:
    print('Result = ',result)
finally:
    print('prgram finished')
    
```
