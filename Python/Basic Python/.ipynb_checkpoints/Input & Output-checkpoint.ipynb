### Output :- 

The print() function is used to display output on the screen.

```python
print('Hello World')
```

##### muliple value

```python
name = 'Alice'
age = 25
print('Name :',name , 'Age :', age)
```

##### using paramter (end & sep)

```python
print('Hello', 'World', sep = "-")
```

```python
print('Hello World' , end = " ")
print('is awesome')
```

##### formated output (using f-string)

```python
name = 'Alice'
age = 24
print(f'My name is {name} and I am {age} years old.')
```

### Input :-

The input() function is used to take user input from the keyboard. It always returns a string.

```python
name = input('Enter a name :')
print('Hello', name)
```

##### taking number input

```python
age = int(input('Enter a age :'))
print(f'Your are {age} years old.')
```

##### taking multiple input

```python
x,y = map(int,input('Enter a two number :').split())
print(f'Sum : {x+y}')
```
