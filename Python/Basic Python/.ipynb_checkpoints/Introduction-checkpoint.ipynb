# Introduction

#### 
Python is a popular programming langauge.

It was created by **Guido Van Rossum** and released in 1991.

It is widly used for web developments, data analysis , artificial intelligence , machine learning and so more.

### Key Features:-

### 
**1.Easy to learn and use -** Pyhton has a simple syntax similar to english.

**2.Interpreted langauge -** Python code is executed line by line , making debugging easier.

**3.Dynamically Typed -** No need declare variable types.

**4.Cross-Platform -** Works on windows,macos and linux.

**5.Large Community support -** A vast number of library and frameworks.
