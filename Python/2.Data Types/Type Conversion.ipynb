##### Type Conversion :-

Type Conversion = changing one data type into another.

##### Types :-

| Type     | Description                | Example       |
| -------- | -------------------------- | ------------- |
| Implicit | Automatic (Python does it) | `int → float` |
| Explicit | Manual (you do it)         | `int()`       |


##### Implicit Type Conversion :-

Python automatically converts data type

```python
a = 10
b = 2.4
c = a+b
print(c)
```

```python
print(type(c))
```

```python
a = 'h'
b = 3

c = a+b
print(c)
```

##### Explicit Type Conversion (Type Casting) :-

You convert manually using functions.

```python
a = 39
float(a)
```

```python
int(39.998)
```
