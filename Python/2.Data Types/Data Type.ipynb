### Data Type :- 

A data type defines the type of value a variable can store.

Data Type = “Type of data” stored in a variable

```python
x = 10      # int data type
x
```

#### Quick Summary :- 

Numeric → Numbers

String → Text

List/Tuple → Ordered data

Set → Unique values

Dictionary → Key-value

#### Python Data Types Table :-

| Category      | Data Type | Example          | Description                         |
| ------------- | --------- | ---------------- | ----------------------------------- |
| 🔢 Numeric    | `int`     | `10`             | Whole numbers                       |
| 🔢 Numeric    | `float`   | `3.14`           | Decimal numbers                     |
| 🔢 Numeric    | `complex` | `2+3j`           | Complex numbers                     |
| 📝 Text       | `str`     | `"Hello"`        | Sequence of characters              |
| 📋 Sequence   | `list`    | `[1, 2, 3]`      | Ordered, mutable (changeable)       |
| 📦 Sequence   | `tuple`   | `(1, 2, 3)`      | Ordered, immutable (not changeable) |
| 🔁 Collection | `set`     | `{1, 2, 3}`      | Unordered, unique values            |
| 🔑 Mapping    | `dict`    | `{"a": 1}`       | Key-value pairs                     |
| ⚖️ Boolean    | `bool`    | `True` / `False` | Logical values                      |
