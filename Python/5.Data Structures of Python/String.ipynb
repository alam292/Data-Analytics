### String :- 

A string in python is a sequence of characters enclosed in quotes are single , double , triple(' ', " ", ''' ''').

Strings are immutable ---> cannot chnage after creation.

Indexed ---> access characters using index.

Iterable ---> can use loops.

#### Indexing :-

Indexing is the process of accessing a single character from a string using its position (index number).

**Key Points:-**

Index start from 0.

Each character has a unique index.

Supports positive and negative indexing.

```python
s = "Python"
```

```python
s[0]
```

```python
print(s[4])
```

```python
print(s[1])
```

```python
print(s[0])   # positive  indexing
print(s[-6])  # negative indexing
```

```python
print(s[-3])
print(len(s[-3]))
```

#### Slicing :- 

Slicing is the process of extracting a substring (part of a string) using a range of indices.

**Syntax:-   string[start : end : step]**

**Key Points :-**

start ---> starting index (included)

end ---> ending index (excluded)

step ---> jump (optinal but by default 1)

```python
s = "Python"
```

```python
print(s[0:6])
```

```python
len(s)
```

```python
print(s[0:5])
```

```python
print(s[0:6:1])
```

```python
print(s[:6:])
```

```python
print(s[0::])
```

```python
print(s[::1])
```

##### Entire string slice :-

No Start, No Index , No Step

```python
print(s[::])
```

```python
print(s[4:6])
```

```python
print(s[0:6:2])
```

```python
print(s[::1])     # positive slicing
print(s[::-1])    # negative slicing
```

```python
print(s[-1:-5:-1])
```

```python
print(s[6:0:-1])     # Stops before index 0 (p is NOT included)
```

#### Looping Through String 

```python
# using for loop

s = 'python'
for x in s:
    print(s)
```

```python
s = 'python'
for x in s:
    print(x)
```

```python
# using index loop

s = 'python'
for i in range(len(s)):
    print(s[i])
```

```python
# using while loop

s = 'python'
i = 0
while i<len(s):
    print(s[i])
    i+=1
```

#### String Repetition :-

The * operator repeats a string multiple times.

```python
s = 'Hi'
print(s*5)
```

```python
s = 'M '
print(s * 5)
```

#### String Concatenation :-

To concatenation or combine, two string you can use the + operator.

```python
a = 'hello'
b = 'python'
c = a + " " + b
print(c)
```

#### formate string :-

we can combine string and numbers by using f-string or the formate() method.

```python
age = 37
txt = f'My  name is John .I am {age} years old.'
print(txt)
```

```python
name = 'Alam'
age = 22
print(f'My Name is {name} . I am {age} years old.')
```

#### Placeholder & Modifiers :-

A placeholder can contain variables, operations, functions and modifiers to formate the value.

```python
price = 93
txt = f'The price is {price} dollar.'
print(txt)
```

```python
price = 82
txt = f'The price is {price : .2f} dollar.'
print(txt)
```

```python
price = 93
txt = f'The price is {price *2} dollar.'
print(txt)
```

#### Python String Methods :-

| Category           | Method         | Description          | Example                 |
| ------------------ | -------------- | -------------------- | ----------------------- |
| 🔤 Case Conversion | `lower()`      | Convert to lowercase | `"HI".lower()` → hi     |
|                    | `upper()`      | Convert to uppercase | `"hi".upper()` → HI     |
|                    | `title()`      | Title case           | `"hello world".title()` |
|                    | `capitalize()` | First letter capital | `"python".capitalize()` |
|                    | `swapcase()`   | Swap case            | `"Hi".swapcase()` → hI  |
|                    | `casefold()`   | Strong lowercase     | `"HELLO".casefold()`    |

| Category                      | Method     | Description         | Example          |
| ----------------------------- | ---------- | ------------------- | ---------------- |
| 🧹 Trimming / Removing Spaces | `strip()`  | Remove both sides   | `" hi ".strip()` |
|                               | `lstrip()` | Remove left spaces  | `" hi".lstrip()` |
|                               | `rstrip()` | Remove right spaces | `"hi ".rstrip()` |

| Category               | Method     | Description                     | Example                  |
| ---------------------- | ---------- | ------------------------------- | ------------------------ |
| 🔍 Searching / Finding | `find()`   | Find index                      | `"hello".find("e")` → 1  |

|                        | `rindex()` | Last index                      | `"hello".rindex("l")`    |
