### List :-

changeable (mutable)

```python
a = [1,2,3,4]
b = ['apple','banana','mango']
c = [1,4,'hello',29.4]
print(a)
print(b)
print(c)
```

##### Accessing list elements`

```python
a = [10,20,30,40]
print(a[0])
```

```python
print(a[-1])
```

##### list slicing

```python
a = [1,2,3,4,5,6]
print(a[1:4])
print(a[2:3])
print(a[0:])
print(a[:6])
print(a[::2])
```

```python
a = [1,2,3,4,5,6]
print(a[1:4])
print(a[2:3:-1])
print(a[::-1])
print(a[:6])
print(a[::-2])
```

##### loop through list

```python
a = [10,20,30,40,50]
for x in a:
    print(x)
```

#### Pyhton List methods :-

| Method               | Description              | Example                                     |
| -------------------- | ------------------------ | ------------------------------------------- |
| `append(x)`          | Adds element at end      | `a=[1,2]; a.append(3) → [1,2,3]`            |
| `extend(iterable)`   | Adds multiple elements   | `a=[1]; a.extend([2,3]) → [1,2,3]`          |
| `insert(i, x)`       | Insert at index          | `a=[1,3]; a.insert(1,2) → [1,2,3]`          |
| `remove(x)`          | Removes first occurrence | `a=[1,2,2]; a.remove(2) → [1,2]`            |
| `pop()`              | Removes last element     | `a=[1,2,3]; a.pop() → [1,2]`                |
| `pop(i)`             | Removes element at index | `a=[1,2,3]; a.pop(1) → [1,3]`               |
| `clear()`            | Removes all elements     | `a=[1,2]; a.clear() → []`                   |
| `index(x)`           | Returns index of element | `a=[1,2,3]; a.index(2) → 1`                 |
| `count(x)`           | Counts occurrences       | `a=[1,2,2]; a.count(2) → 2`                 |
| `sort()`             | Sorts list ascending     | `a=[3,1,2]; a.sort() → [1,2,3]`             |
| `sort(reverse=True)` | Sort descending          | `a=[1,2,3]; a.sort(reverse=True) → [3,2,1]` |
| `reverse()`          | Reverses list            | `a=[1,2,3]; a.reverse() → [3,2,1]`          |
| `copy()`             | Returns copy of list     | `a=[1,2]; b=a.copy()`                       |
