# Practices Questions

### String Questions

##### 1. Count vowels in a string

Methods used: lower(), count()

```python
s = 'Hello World'
count = 0
for x in s.lower():
    if x in 'aeiou':
        count+=1
print(count)
```

##### 2. Reverse a string

Methods used: slicing

```python
s = 'Python'
print(s[::-1])
```

##### 3. Check palindrome

Methods used: slicing, lower(), replace()

```python
s = 'Madam'
if s == s.lower()[::-1]:
    print('Palidrome')
else:
    print('Not Palidrome')
```

##### 4. Count words in a string

Methods used: split()

```python
s = "I Love Python Programming"
print("Total words : " ,len(s.split()))
```

##### 5. Remove spaces from string

Methods used: replace()

```python
s = "Hello World"
print(s.replace(" ",""))
```

##### 6. Convert string to uppercase & lowercase

Methods used: upper(), lower()

```python
s = "Python"
print(s.upper())
print(s.lower())
```

##### 7. Find frequency of each character

Methods used: count()

```python
s = 'programming'
for x in set(s):
    print(x ," : ",s.count(x))
```

##### 8. Check if string contains substring

Methods used: in, find()

```python
s = "Hello Python"
if "Python" in s:
    print('Found')
print(s.find('Python'))
```

##### 9. Replace vowels with *

Methods used: replace()

```python
s = 'Hello World'
for s in 'aeiouAEIOU':
    s= s.replace(s,'*')
print(s)
```

##### 10. Capitalize first letter of each word

Methods used: title(), capitalize()

```python
s = 'hello world'
print(s.title())
print(s.capitalize())
```

##### 11. Remove duplicate characters

Methods used: set(), join()

```python
s = "programming"
result = ''.join(set(s))
print(result)
```

##### 12. Join list into string

Methods used: join()

```python
s = ['Python','is','easy']
print(' '.join(s))
```

##### 13. Split string by comma

Methods used: split()

```python
a = 'apple,banana,mango'
print(a.split(','))
```

##### 14.Find first non-repeating character

Methods used: count()

```python
a = "aabbcdde"
for x in a:
    if a.count(x) == 1:
        print("First Non-Repeating :",x)
        break
```

##### 15.Check anagram

Methods used: sorted()

```python
s1 = "listen"
s2 = "silent"
if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Non Anagram")
```
