### Tuple :-

A tuple is a collection of items that is :

Ordered 

Immutable (cannot be changed)

Allows duplicate values

```python
t = (1,2,3,4)
print(t)
```

```python
print(type(t))
```

```python
t = (1,3.5,3,'hi',True,'apple')
print(t)
```

```python
t = ()
print(t)
print(type(t))
```

```python
t = (,)
print(type(t))
print(t)
```

```python
t = (4,)
print(t)
print(type(t))
```

```python
t = (10,20,30)
print(t[0])
print(t[-1])
```

```python
t = (1,2,3,4,5,6)
print(t[:])
print(t[0:])
print(t[2:5])
```

```python
t = (1,2,3)
t[0] = 6
print(t)
```

```python
t = (1,2,3,4,5)
for i in t:
    print(i)
```

```python
print(len(t))
```

##### convert tuple

```python
# tuple ---> list

t = (1,2,3,4)
l = list(t)
print(l)
```

```python
# list ---> tuple

l = [1,2,3,4]
t = tuple(l)
print(t)
```

##### Tuple Packing and Unpacking

```python
# packing

t = 1,2,3,4

# unpacking

a,b,c,d = t
print(a,b,c,d)
```

### Types :-

| Method    | Description       | Example      |
| --------- | ----------------- | ------------ |
| `count()` | Count occurrences | `t.count(2)` |
| `index()` | Find index        | `t.index(3)` |


```python
t = (1, 2, 2, 3)

print(t.count(2))   # 2
print(t.index(3))   # 3
```

```python
t = (1,2,3,2,4,3,2)
print(t.count(2))
```

```python
t = (1, 2, 3)

print(t.count(5))
```

```python
t = (10, 20, 30, 40)

print(t.index(30))
```

```python
t = (1, 2, 2, 3)

print(t.index(2))
```

```python
t = (1, 2, 2, 3)

print(t.index(2,2))
```
