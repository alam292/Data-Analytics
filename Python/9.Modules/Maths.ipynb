### Maths :-

The math module provides mathematical functions and constants used in calculations.

```python
import math
```

### Built-in Math functions :-

#### min() --> lower value

```python
x = min(5,2,4,3)
x
```

#### max() --> higher value

```python
x = max(2,4,3,5,1)
x
```

#### fabs()/abs() --> returns the absolution(positive) value of a number.

```python
print(abs(-103.4))
```

#### sqrt() --> square root

```python
print(math.sqrt(16))
```

#### ceil() --> round up

```python
print(math.ceil(29.4))
```

#### floor() --> round down

```python
print(math.floor(28.3))
```

#### pow() --> power

```python
print(math.pow(2,3))
```

#### factorial() --> factorial (!5 = 5 x 4 x 3 x 2 x 1)

```python
print(math.factorial(5))
```

#### gcd() --> greater common divisor

```python
print(math.gcd(12,18))
```

#### lcm() --> least common multiple

```python
print(math.lcm(12,18))
```
