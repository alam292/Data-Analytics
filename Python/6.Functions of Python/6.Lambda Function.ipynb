### What is a Lambda Function ?

A lambda function is a small anonymous (unnamed) function wriiten in a single line.

Used for short, simple operations without defining a full function using **def**.

Syntax :- **lambda arguments : expression**

No function name

No return keyword

Automatically return result

```python
square = lambda x : x*x
print(square(3))
```

#### Lambda vs Normal Function

| Feature     | Lambda             | def Function |
| ----------- | ------------------ | ------------ |
| Name        | No                 | Yes          |
| Size        | One-line           | Multi-line   |
| Complexity  | Simple             | Complex      |
| Reusability | Low                | High         |
| Readability | Low (complex case) | High         |


##### Mutliple arguments example

```python
add = lambda a,b :a+b
print(add(3,5))
```

```python
even_odd = lambda n : n%2==0
print(even_odd(9))
```
