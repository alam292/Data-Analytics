### Built-ib Functional Tools :-

These functionals help you:

Write clean and short code.

Apply operations on collections(list, tuple,etc).

Follow functional programming style.

##### 1.map() :-

Applies a function to every element of sn iterable (list,tuple,etc).

**Syntax :- map(function, iterable)**

**Use Case :-**
Data transformation

Mutliply,Square,convert data

```python
nums = [1,2,3,4,5]

result = list(map(lambda x : x*2 , nums))
print(result)
```

```python
# without lambda

def square(x):
    return x*x
print(square(4))
```

```python
# without lambda

def square(x):
    return x*x
print(list(map(square,nums)))
```

##### 2.filter() :- 

filter elements based on  a condition.

**Syntax :-filter(function, iterable)**

Function must return true or false.

**Use Case :-**

Filtering data

Conditions (even, odd, > value, etc.)

```python
nums = [1,2,3,4,5]
even = list(filter(lambda x : x%2==0, nums))
print(even)
```

```python
# without lambda

def is_even(n):
    return n%2==0

print(list(filter(is_even,nums)))
```

##### 3.reduce() :-

Applies a function cumulatively to elements ----> returns single value.

Must import from *functools*

**Syntax:-**   *from functools import reduce*

**reduce(function, iterable)**

**Use Case :-**

Sum of list 

Product of list

Aggregation

```python
from functools import reduce
nums = [1,2,3,4]
result = reduce(lambda x, y : x+y, nums)
print(result)
```

#### Comparison :- 

| Feature   | map()     | filter() | reduce()          |
| --------- | --------- | -------- | ----------------- |
| Purpose   | Transform | Filter   | Aggregate         |
| Output    | Iterable  | Iterable | Single value      |
| Condition | No        | Yes      | No                |
| Import    | No        | No       | Yes (`functools`) |


map() → Modify data

filter() → Select data

reduce() → Combine data

#### Real-life example

```python
from functools import reduce

nums = [10, 20, 30, 40]

# Step 1: Increase values
mapped = list(map(lambda x: x + 5, nums))

# Step 2: Filter > 25
filtered = list(filter(lambda x: x > 25, mapped))

# Step 3: Sum all
result = reduce(lambda x, y: x + y, filtered)

print(result)
```
