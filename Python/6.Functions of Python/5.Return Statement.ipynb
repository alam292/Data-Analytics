### Return Statement :-

The return statement is used to :

Send a value back from a function.

Stop execution of the function.

**Note :-** Once **return** is executed,the function ends immediately.

##### 1.Returning Single Value :- 

Function returns only one value.

```python
def add(a,b):
    return a+b

result = add(3,5)
print(result)
```

##### 2.Returning Multiple Value:-

Python allows returning multiple values at once.

Actually it returns a tuple.

```python
def calc(a,b):
    return a+b , a-b

sum_val, diff_val = calc(10,4)
print(sum_val)
print(diff_val)
```

##### 3.Returning None :-

If a function does not retutn anything, it returns None by default.

```python
def greet():
    print('Hello')

result = greet()
print(result)
```

##### Explicit None

```python
def check(num):
    if num > 0:
        return "Positive"
    else:
        return None

print(check(-5))
```

```python
def student_result(marks):
    total = sum(marks)
    avg = total / len(marks)
    return total, avg

t, a = student_result([80, 90, 85])

print("Total:", t)
print("Average:", a)
```
