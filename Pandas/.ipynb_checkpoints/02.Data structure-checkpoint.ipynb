# Data Structure

### Series(1-D)

A Series is a one-dimensional labeled array in Pandas.

Think of it like a single column in Excel.

**Key Features:-**

Has data + index (labels)

Can store any data type (int, float, string)

Similar to NumPy array but with labels

```python
import pandas as pd
```

```python
data = [10,20,30,40,50]
s = pd.Series(data)
print(s)
```

```python
data2 = ['apple','banana','cherry','date']
a = pd.Series(data2)
print(a)
```

```python
type(s)
```

```python
type(a)
```

```python
len(a)
```

```python
len(s)
```

##### students marks (single subject)

```python
marks = pd.Series([92,67,88],index = ['Aman','Riya','John'])
print(marks)
```

##### stock price of one company

```python
prices = pd.Series([100,102,108],index =['Day1','Day3','Day2'])
print(prices)
```

### DataFrame

A DataFrame is a two-dimensional table (rows + columns).

Think of it like a full Excel sheet or SQL table.

**Key Features:-**
    
Multiple columns (each column = Series)
    
Different data types allowed
    
Indexed rows and labeled columns

```python
data = {'Name':['Aman','Riya','Saniya']}
print(data)
```

```python
type(data)
```

```python
len(data)
```

```python
data = {'Name':['Aman','Riya','Saniya'],'Marks':[85,93,76],'City':['Delhi','Mumbai','Patna']}
df = pd.DataFrame(data)
print(df)
```

##### student dataset

```python
df = pd.DataFrame({'Name':['Aman','Riya','Saniya'],'Marks':[85,93,76],'Branch':['ME','CS','ECE']})
print(df)
```

##### healthcare data

```python
df = pd.DataFrame({'PatientID':[1,2,3],'Age':[45,50,29],'Diabetes':[1,1,0]})
print(df)
```
