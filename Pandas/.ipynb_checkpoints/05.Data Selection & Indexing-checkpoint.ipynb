# Data Selection & Indexing

```python
import pandas as pd
```

```python
df = pd.read_csv('India_Ecommerce_Unclean_Dataset.csv')
```

```python
df
```

```python
df.info()
```

### 1.Selection columns

##### single column ----> returns a series(1D)

```python
df['Customer_Phone']
```

```python
df['Return_Status']
```

##### multiple columns ----> returns a dataframe (2-D)

```python
df[['Customer_Name','Discount_Percent']]
```

```python
df[['Customer_Name','Discount_Percent','Payment_Method','Payment_Status']]
```

### 2.Row Selection 

##### loc[ ] -----> label-based selection , uses index labels(names)

```python
df.loc[0]
```

```python
df.loc[5]
```

```python
type(df.loc[5])
```

```python
df.loc[0:5]
```

```python
type(df.loc[0:5])
```

```python
df.loc[4:13]
```

```python
df.loc[2:20:2]
```

```python
a = df.loc[[2,10,5],['Order_ID','Customer_Name','Customer_City','Tax_Percent','Shipping_Method']]
```

```python
a
```

```python
type(a)
```

```python
df.loc[:, ['Customer_Name','Customer_City','Tax_Percent']]
```

##### iloc [ ] ---> Position-based selection , uses index position (numbers)

```python
df.iloc[0]
```

```python
df.iloc[0:5]
```

```python
df.iloc[4:11:3]
```

```python
df.iloc[[1,4,2],5]
```

```python
df.loc[[1,4,2],5]
```

```python
df.iloc[[2,5,10], :]
```

```python
df.loc[[2,5,10], :]
```

```python
df.iloc[:, [3, 5, 9]]
```

### 3.Boolean Filtering

```python
df.info()
```

```python
df['Quantity'] > 5
```

```python
df[df['Quantity'] > 5]
```

```python
df['Order_Date'] > '2024/01/01'
```

```python
df['Customer_City'] == 'Chenai'
```

```python
df[df['Customer_City'] == 'Chenai']
```
