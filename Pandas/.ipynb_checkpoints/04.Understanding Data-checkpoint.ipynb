# Understanding Data

Data Understanding in Pandas is the first and most important step in Exploratory Data Analysis (EDA).
It helps you quickly inspect, summarize, and get insights about your dataset 
before applying any transformations or models.

### Load data 

```python
import pandas as pd
```

```python
df = pd.read_csv('India_Ecommerce_Unclean_Dataset.csv')
df
```

### view the data

##### first and last rows

```python
df.head()
```

```python
df.tail()
```

##### random sample

```python
df.sample(5)
```

### Check shape & structure

##### shape (rows, columns)

```python
df.shape
```

##### columns name

```python
df.columns
```

##### data types

```python
df.dtypes
```

### Dataset Summary Info

**Gives:**

Total rows

Column names

Data types

Non-null values

**Help identify:**

Missing values count

wrong data types

```python
df.info()
```

### Statistical summary

##### numerical columns

```python
df.describe()
```

##### categorical columns

```python
df.describe(include = 'O')
```

##### all columns

```python
df.describe(include = 'all')
```

### Row Index

```python
df.index
```

### Missing Values Analysis

```python
df.isnull()
```

```python
df.isnull().sum()
```

##### percentage of missing values

```python
(df.isnull().sum() / len(df)) * 100
```

### Duplication Data

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

### Unique Values

```python
df.nunique()
```

##### unique values in a specific column

```python
df['Region'].unique()
```

```python
df['Tax_Percent'].unique()
```

### Value Counts (Categorical Understanding)

```python
df['Customer_City'].value_counts()
```

```python
df['Customer_City']['Delhi'].value_counts()
```

### Coorelation (Relationships)

```python
df.corr()
```
