### Check & Correct Data Types

```python
df.dtypes
```

```python
df['Customer_Phone'] = df['Customer_Phone'].astype('string')
```

```python
cols = ['Region', 'Category', 'Sub_Category','Payment_Method','Payment_Status','Shipping_Method','Return_Status']
df[cols] = df[cols].astype('category')
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)
```

```python
df.dtypes
```

```python
import pandas as pd
import numpy as np

# Fix dates
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"], errors="coerce")

# Fix price (remove ₹, commas)
df["Unit_Price_INR"] = pd.to_numeric(
    df["Unit_Price_INR"].str.replace(r"[₹,]", "", regex=True), errors="coerce"
)

# Fix pincode (12345.0 → "012345")
df["Customer_Pincode"] = df["Customer_Pincode"].fillna(0).astype(int).astype(str).str.zfill(6)
df["Customer_Pincode"] = df["Customer_Pincode"].replace("000000", np.nan)

# Fix phone (keep digits only)
df["Customer_Phone"] = df["Customer_Phone"].str.replace(r"\D", "", regex=True)

print(df.dtypes)
```

### Handel Outliers

Outliers are extreme values that are very different from the rest of the data.

10, 12, 15, 18, 20, 500

Here 500 is an outlier.

**Outliers mainly exist in numerical columns (int and float).**

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

num_cols
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,10))

plotnumber = 1

for col in num_cols:
    
    plt.subplot(3,3,plotnumber)
    sns.boxplot(y=df[col])
    plt.title(col)
    
    plotnumber += 1

plt.tight_layout()
plt.show()
```

```python
for col in num_cols:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    
    print(f"{col} : {len(outliers)} outliers")
```

```python
for col in num_cols:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]
    
    if len(outliers) > 0:
        print(f"{col} has outliers")
```
